```python
import tensorflow as tf
import numpy as np
import os
```

```python
from tensorflow.keras.applications.mobilenet_v3 import preprocess_input
```

```python
model=tf.keras.models.load_model("trained_retinal_model.h5")
```

```python
model.summary()
```

```python
img_path=r"..\data\retina\test\DME\DME-30521-17.jpeg"
img=tf.keras.utils.load_img(img_path,target_size=(224,224))
img
```

```python
x=tf.keras.utils.img_to_array(img)
x=np.expand_dims(x,axis=0)
x.shape
x=preprocess_input(x)
```

```python
x.shape
```

```python
pred=model.predict(x)
```

```python
pred
```

```python
class_name=['CNV','DME','DRUSEN','NORMAL']
```

```python
result_index=np.argmax(pred)
print("Model prediciton:",class_name[result_index])
```
